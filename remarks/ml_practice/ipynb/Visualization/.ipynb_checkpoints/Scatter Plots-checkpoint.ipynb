```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
insurance_filepath= "../../datasets/insurance.csv"
insurance_data = pd.read_csv(insurance_filepath)
```

```python
insurance_data.head()
```

### 산점도(Scatter Plot)

```python
sns.scatterplot(x=insurance_data.bmi, y = insurance_data.charges)
```

### Regression Line
- bmi지수가 보험료와 연관이 있다.그니까 회귀곡선을 한번 그려보자

```python
sns.regplot(x = insurance_data.bmi, y = insurance_data.charges)
```

### 세개의 변수에 대한 관계를 그릴 수 있다.
- 흡연여부가 bmi와 보험료에 미치는 영향 등

```python
sns.scatterplot(x=insurance_data.bmi, y=insurance_data.charges, hue=insurance_data.smoker)
```

### 두개의 회귀 선을 그려보자

```python
sns.lmplot(x="bmi", y="charges", hue="smoker", dabta = insurance_data)
```

### Categorical Scatter Plot
- 산점도를 그릴때는 보통 두 연속 변수의 관계를 보려고 하는 목적.
- categorical Variable에도 쓸 수 있다.
    - 이 때는 swarmplot으로 그린다.

```python
sns.swarmplot(x=insurance_data["smoker"], y=insurance_data["charges"])
```
