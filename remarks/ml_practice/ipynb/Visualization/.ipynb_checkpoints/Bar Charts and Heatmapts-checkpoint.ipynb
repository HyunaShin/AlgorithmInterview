### 라이브러리, 데이터셋 불러오기

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
flight_path = "../../datasets/flight_delays.csv"
flight_data = pd.read_csv(flight_path,index_col="Month", parse_dates=True)
flight_data.head()
```

```python
flight_data
```

### Bar Chart

```python
#그래프 너비랑 크기 정하기
plt.figure(figsize=(10,6))
#제목 정하기
plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month")

#Spirit Airlines의 월별 항공 지연을 시각화하는 바차트 보여주기
sns.barplot(x = flight_data.index, y=flight_data["NK"])

#수직 축에 라벨 추가하기
plt.ylabel("Arrival Delay(in minutes)")
```

### Heat Map

```python
plt.figure(figsize=(14,7))
plt.title("Average Arrival Delay for Each Airline, by Month")
#각각의 항공사마다 월별 평균 지연 시간을 보여주기
#annot : 셀에 숫자도 보여줄지말지
sns.heatmap(data=flight_data, annot=True)

plt.xlabel("Airline")
```
