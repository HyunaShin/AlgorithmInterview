```python
melbourne_data = melbourne_data.dropna(axis = 0)
```

### 피쳐(X), 타겟(Y) 선정하기

```python
y = melbourne_data.Price
```

```python
melbourne_data_features = ["Rooms", "Bathroom", "Landsize", "Lattitude", "Longtitude"]
x = melbourne_data[melbourne_data_features]
```

### 모델 빌딩하기
- Define: What type of model will it be? A decision tree? Some other type of model? Some other parameters of the model type are specified too.
- Fit: Capture patterns from provided data. This is the heart of modeling.
- Predict: Just what it sounds like
- Evaluate: Determine how accurate the model's predictions are.

### Random State?
- 머신러닝 모델은 모델 학습시 랜덤을 부여하기도 함
- random_state에 숫자를 부여하면, 여러번 실행시켜도 같은 결과를 뱉아냄
- 어느 번호를 써도 된다. 모델은 고른 숫자에 크게 의미를 두고 의존을 갖진 않음.

```python
from sklearn.tree import DecisionTreeRegressor
#실행할 때 마다 같은 결과를 뱉아내게, random_state는 1로 해두셈
melbourne_model = DecisionTreeRegressor(random_state=1)
#fit model
melbourne_model.fit(x,y)
```

### 예측하기
- 원래는 집값을 모르는 집에 대해서 예측해야되지만, 예측하는걸 좀 보려고 이미 학습한 데이터를 넣어봄

```python
print("Making predictions for the following 5 houses:", end = "\n\n")
print(x.head(), end = "\n\n")
print("The predictions are", end = " ")
print(melbourne_model.predict(x.head()))
```

# Model Validation
There are many metrics for summarizing model quality, but we'll start with one called Mean Absolute Error (also called MAE). Let's break down this metric starting with the last word, error.

The prediction error for each house is:

error=actual−predicted

So, if a house cost $150,000 and you predicted it would cost $100,000 the error is $50,000.

With the MAE metric, we take the absolute value of each error. This converts each error to a positive number. 

##### On average, our predictions are off by about X.

```python
from sklearn.metrics import mean_absolute_error

predicted_home_prices = melbourne_model.predict(x)
mean_absolute_error(y, predicted_home_prices)
```

# The Problem with "In-Sample" Scores
- 방금 계산한 스코어는 in-sample 점수다.
- 샘플에서 학습을 시켰을때, 사실은 문색깔은 상관이 없는데 학습시킨 데이터에서는 초록 문이 매우 비쌌다고 치자 
- 이 패턴은 학습 데이터 때문에 생긴 문제니까. 트레이닝 데이터에서는 맞아보일 수도 있다.
- 새 데이터에서 예측을 할때는 잘 안먹는다. 실제로 적용했을 때 부정확할 수 있다.
- 모델의 실제 들어오는 값은, 새로운 데이터이므로 모델을 만들때 안썼던 데이터로 성능을 측정 해 보자.
- 가장 정석은 모델을 빌딩할 때 데이터 몇개를 제외하고 시키고, 못봤던 데이터만 테스트 하는 것.
- 이러한 데이터를 validation Data라고 한다.

```python
from sklearn.model_selection import train_test_split

train_X, val_X, train_y, val_y= train_test_split(x,y, random_state=0)
melbourne_model = DecisionTreeRegressor()
melbourne_model.fit(train_X, train_y)
val_predictions = melbourne_model.predict(val_X)
```

```python
print(mean_absolute_error(val_y, val_predictions))
```
